```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the database
db_path = "C3S_V2.db"
con = sqlite3.connect(db_path)
cur = con.cursor()

# Query data (adjust the query based on your table structure and analysis goals)
query = "SELECT * FROM master"  # Replace 'your_table_name' with the actual table name
df = pd.read_sql_query(query, con)

# Close the database connection
con.close()

grouped = df.groupby("adduct")
grouped.head()
```

```python
# Display the DataFrame to check the data (optional)
print(df.head())

print(df.describe())  # Gives basic statistics (mean, std, min, etc.)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

common_adduct = ["[M+H]+", "[M+Na]+", "[M-H]-", "[M+NH4]+", "[M+K]+",
                 "[M+H-H2O]+", "[M+HCOO]-", "[M+CH3COO]-", "[M+Na-2H]-"]

df_filter = df[df['adduct'].isin(common_adduct)]
columns_to_include = [col for col in df.columns if col not in ["z", "mass"]]

pair_plot_all = sns.pairplot(df[columns_to_include], diag_kind="kde", hue="adduct")
pair_plot_all.figure.suptitle("Facet Grid of Every Feature and Data Point in C3SDB_V2", y=1.02)

pair_plot_filter = sns.pairplot(df_filter[columns_to_include], diag_kind="kde", hue="adduct")
pair_plot_filter.figure.suptitle("Facet Grid of Common Adducts in C3SDB_V2", y=1.02)

plt.show()
```

```python
from scipy.stats import linregress

regression_results={}
# Perform linear regression for each group
for name, group in grouped:
    # Extract x and y features for regression
    x = group['mz']
    y = group['ccs']
    
    # Perform linear regression
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    
    # Compute R^2 value
    R_squared = r_value ** 2
    
    # Save results
    regression_results[name] = {
        'slope': slope,
        'intercept': intercept,
        'r_value': r_value,
        'R_squared': R_squared,
        'p_value': p_value,
        'std_err': std_err
    }
    
    # Plot each group with regression line
    plt.figure(figsize=(6, 4))
    plt.scatter(x, y, label=f'{name} data', alpha=0.5)
    plt.plot(x, slope * x + intercept, color="red", label='Fit line')
    plt.xlabel('mz')
    plt.ylabel('ccs')
    plt.title(f'Linear Regression for Adduct: {name}\n$R^2$ = {R_squared:.4f} $n$ = {len(group)}')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
# Perform linear regression for each group
regression_results={}
selected_adducts = ["[M+H]+", "[M+HCOO]-", "[M+H-H2O]+"]

for name, group in grouped:
    if name in selected_adducts:

        group = group.dropna(subset=['mz', 'polarization'])
        
        # Check if there's enough data left after dropping missing values
        if group.empty or group['polarization'].nunique() <= 1:
            print(f"Skipping adduct {name} due to insufficient data after cleaning.")
            continue

        # Extract x and y features for regression
        x = group['mz']
        y = group['polarization']
        
        # Perform linear regression
        slope, intercept, r_value, p_value, std_err = linregress(x, y)
        
        # Compute R^2 value
        R_squared = r_value ** 2
        
        # Save results
        regression_results[name] = {
            'slope': slope,
            'intercept': intercept,
            'r_value': r_value,
            'R_squared': R_squared,
            'p_value': p_value,
            'std_err': std_err
        }
        
        # Plot each group with regression line
        plt.figure(figsize=(6, 4))
        plt.scatter(x, y, label=f'{name} data', alpha=0.5)
        plt.plot(x, slope * x + intercept, color="red", label='Fit line')
        plt.xlabel('mz')
        plt.ylabel('polarization')
        plt.title(f'Linear Regression for Adduct: {name}\n$R^2$ = {R_squared:.4f} $n$ = {len(group)}')
        plt.legend()
        plt.tight_layout()
        plt.show()
```
